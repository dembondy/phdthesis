```python
import scipy.io as spio
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
rms_data = spio.loadmat('rms_all.mat')
rms = np.transpose(rms_data['RMS'][::-1,:])
availability = np.linspace(70,100,7)
noise = np.linspace(0.3,0.0,7)
sh = np.shape(rms)
rms2 = rms.copy()
for i in range(sh[0]):
    for j in range(sh[1]):
        rms2[i,j]=round(rms[i,j],1)
        
print(rms)
print(rms2)
```

```python
plot1 = plt.figure()
sns.set(font="Arial",context = "paper", font_scale = 1.5)
sns.heatmap(rms2,cmap='viridis',xticklabels=availability,yticklabels=noise,annot=True,fmt='g')
plt.xlabel('Portfolio availability [%]')
plt.ylabel(r'Process noise [$\sigma$]')
plt.show()
#plot1.savefig("heatmap.pdf")
```
